```python
import pandas as pd
import numpy as np

csv_list = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 
            'August', 'September', 'October', 'November', 'December', 'January 2021', 
            'February 2021']

for i, csv in enumerate(csv_list):
    twitter_month = pd.read_csv('Twitter Data/Twitter - ' + csv + '.csv')
    
    if i == 0:
        twitter_full = twitter_month
        
    else:
        twitter_full = pd.concat([twitter_full, twitter_month])
    
```

```python
twitter_full = twitter_full.reset_index(drop = True)

# Remove all columns with the name 'promoted'
twitter_full = twitter_full.loc[:, ~twitter_full.columns.str.contains('promoted')]

twitter_full.to_csv('Twitter_Full_Data.csv')
```

```python
# Scrapy crawlers DON'T WORK on Jupyter. Rather, they need you to restart the kernel 
# every time you want to run the notebook. 

import scrapy
from scrapy.crawler import CrawlerProcess

url = 'https://pokemondb.net/pokedex/national'

# Create our spider
class PokemonScraper(scrapy.Spider):
    name = 'pokespider'
    def start_requests(self):
        yield scrapy.Request(url = url, callback = self.parse)
        
    def parse(self, response):
        # Note: If we want to return our results, we'll need to append it to a list
        # or dictionary OUTSIDE of the spider.
        links = response.xpath('//h2[@id="gen-8"]/following::div/div[@class="infocard "]/span[@class="infocard-lg-img"]//@href').extract()
        for link in links:
            yield response.follow(url = link, callback = self.parse_moves)
            
    def parse_moves(self, response):
        # Name of pokemon - note that we're looking at all h1 text inside the Grookey
        # Note that extract returns a list, while extract_first returns a string.
        name = response.css('h1::text').extract_first()
        # Level up section
        levelup_heading = response.xpath('//div[@class="grid-col span-lg-6"][1]')
        # Move names
        levelup_moves = levelup_heading.css('td.cell-name ::text').extract()
        move_dict[name] = levelup_moves
        
move_dict = {}        

process = CrawlerProcess()

process.crawl(PokemonScraper)

process.start()
```

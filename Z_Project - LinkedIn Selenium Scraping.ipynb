```python
# Grab the data that we're interested in. That's the post by post data in the bottom 
# half of the screen. 

# Info we want:
p_content = []
p_link = []
p_date = []
p_impressions = []
p_clicks = []
p_CTR = []
p_reactions = []
p_comments = []
p_shares = []
p_engagement_rate = []

for i in range(36):
    # wait for posts to load
    time.sleep(5 + random.randint(1,5))
    
    # grab the following information from the 10 posts on the page
    titles = driver.find_elements_by_xpath('//div[@class="break-words lt-line-clamp lt-line-clamp--multi-line ember-view"]')
    links = driver.find_elements_by_xpath('//div[@class="break-words lt-line-clamp lt-line-clamp--multi-line ember-view"]/parent::*')
    dates = driver.find_elements_by_xpath('//td[@class="org-update-engagement-table__table-cell text-align-right"][1]')
    impressions = driver.find_elements_by_xpath('//td[@class="org-update-engagement-table__table-cell text-align-right"][2]')
    clicks = driver.find_elements_by_xpath('//td[@class="org-update-engagement-table__table-cell text-align-right"][4]')
    CTR = driver.find_elements_by_xpath('//td[@class="org-update-engagement-table__table-cell text-align-right"][5]')
    reactions = driver.find_elements_by_xpath('//td[@class="org-update-engagement-table__table-cell text-align-right"][6]')
    comments = driver.find_elements_by_xpath('//td[@class="org-update-engagement-table__table-cell text-align-right"][7]')
    shares = driver.find_elements_by_xpath('//td[@class="org-update-engagement-table__table-cell text-align-right"][8]')
    engagement_rate = driver.find_elements_by_xpath('//td[@class="org-update-engagement-table__table-cell text-align-right"][10]')

    for title in titles:
        p_content.append(title.text)

    for link in links:
        p_link.append(link.get_attribute('href'))

    for date in dates:
        p_date.append(date.text)

    for impression in impressions:
        p_impressions.append(impression.text)

    for click in clicks:
        p_clicks.append(click.text)

    for value in CTR:
        p_CTR.append(value.text)

    for reaction in reactions:
        p_reactions.append(reaction.text)

    for comment in comments:
        p_comments.append(comment.text)

    for share in shares:
        p_shares.append(share.text)

    for value in engagement_rate:
        p_engagement_rate.append(value.text)
    
    # linger for a random amount of time
    time.sleep(10 + random.randint(1,5))
    
    # Might need to write something that forces the scroll all the way to the end
    driver.find_element_by_xpath('//button[@aria-label="Next"]').click()
```

```python
data = pd.DataFrame(columns = ['post', 'link', 'date_uploaded', 'impressions', 
                               'clicks', 'click_through_rate', 'reactions', 
                               'comments', 'shares', 'engagement_rate'])

data['post'] = p_content
data['link'] = p_link
data['date_uploaded'] = p_date
data['impressions'] = p_impressions
data['clicks'] = p_clicks
data['click_through_rate'] = p_CTR
data['reactions'] = p_reactions
data['comments'] = p_comments
data['shares'] = p_shares
data['engagement_rate'] = p_engagement_rate

data
```

```python
# Create a dataframe that will hold ALL the data. Do the initial step of creating the data_full
# dataframe. 
# data_full = data

# From here, append the new data df to this one.
data_full = pd.concat([data_full, data])
# Reset and drop the index (pandas doesn't automatically reindex when concatenating)
data_full = data_full.reset_index(drop = True)
data_full
```

```python
# Explanations of each metric

# Impressions = +1 when the update is on screen for 300ms, or when clicked
# Clicks = Clicks on the update
# CTR = Clicks/Impressions
# Reactions = total number of reactions to the update
# Comments/Shares = self explanatory
# Engagement Rate = (Clicks+Reactions+Comments+Shares) / Impressions
# data_full = data_full.reset_index(drop = True)
data_full.to_csv('LinkedIn_Data.csv')
```

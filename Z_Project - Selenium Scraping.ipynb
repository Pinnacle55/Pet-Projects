```python
# Let's try to get all our data (24 pages, October 2019)
# Seems to have worked
details_links = []
analytics_links = []
date_posted = []
vid_titles = []
for i in range(24):
    time.sleep(5 + random.randint(1,5))
    
    # Tried to scroll down to get all the values, but it didn't work
    #     html = driver.find_element_by_tag_name('html')
    #     html.send_keys(Keys.END)

    #     time.sleep(5 + random.randint(1,5))
    
    # Hover over each video to get the link to appear - hopefully this works
    hover_elements = driver.find_elements_by_xpath('//div[@class="style-scope ytcp-video-list-cell-video right-section"]')

    for element in hover_elements[5:]:
        time.sleep(0.5 + random.random())
        hover = ActionChains(driver).move_to_element(element)
        hover.perform()
    
    details = driver.find_elements_by_xpath('//*[@aria-label="Details"]/parent::*')
    # analytics = driver.find_elements_by_xpath('//*[@aria-label="Analytics"]/parent::*')
    dates = driver.find_elements_by_xpath('//div[@id="row-container"]/div[@role="cell"][5]')
    titles = driver.find_elements_by_xpath('//a[@id="video-title"]')
    
    
    for link in details:
        details_links.append(link.get_attribute('href'))
        
    #     for link in analytics:
    #         analytics_links.append(link.get_attribute('href'))
    
    for date in dates:
        date_posted.append(date.text.split("\n")[0])
    
    for title in titles[5:]:
        vid_titles.append(title.text)
    # time.sleep(2)
    # Move to the next page
    driver.find_element_by_id("navigate-after").click()
```

```python
links = pd.DataFrame(columns = ['v_link', 'v_title', 'date_uploaded'])

links['v_link'] = details_links
links['v_title'] = vid_titles
links['date_uploaded'] = date_posted
# df = pd.DataFrame(columns = ['title', 'description'])

links.to_csv('Video_Links_and_Dates.csv')
```
